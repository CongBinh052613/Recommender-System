# Popularity Based Recommender System



```python
ratings_with_books = ratings.merge(books, on='ISBN')
ratings_with_books.head()
```

##### Lọc ra những User nào đánh giá trên 50 phim và những bộ phim nào được hơn 200 đánh giá

```python
num_ratings_df = ratings_with_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings_df.rename( columns ={'Book-Rating':'num_ratings'}, inplace=True)
num_ratings_df
```

```python
avg_ratings_df = ratings_with_books.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings_df.rename( columns ={'Book-Rating':'avg_ratings'}, inplace=True)
avg_ratings_df
```

```python
popular_df = num_ratings_df.merge(avg_ratings_df, on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popular_df=popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','avg_ratings','num_ratings']]
```

```python
popular_df
```

# Collaborative Filtering Based Recommender System

```python
# lấy ra sách có số lượng rating lớn hơn 200
x = ratings_with_books.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_books[ratings_with_books['User-ID'].isin(padhe_likhe_users)]
filtered_rating
```

```python
# lấy ra người dùng có trên 50 lượt đánh giá phim
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_rating = ratings_with_books[ratings_with_books['Book-Title'].isin(famous_books)]
```

```python
final_rating
```

```python
pt = final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
np.where(pt.index=='Zoya')[0][0]
```

```python
sorted(list(enumerate(similarity_scores[0])),key= lambda x: x[1], reverse=True)[1:6]
```

```python
print(popular_df[popular_df['Book-Title']=='1984']['Book-Author'].values)
```

```python
def recommend(name):
    index = np.where(pt.index==str(name))[0][0]
    similarity = sorted(list(enumerate(similarity_scores[index])),key= lambda x: x[1], reverse=True)[1:9]
    recommend = []
    for i in similarity:
        lists=[]
        x = pt.index[i[0]]
        book_temp=books.drop_duplicates('Book-Title')
        lists.extend(list(book_temp[book_temp['Book-Title']==pt.index[i[0]]]['Book-Title'].values))
        lists.extend(list(book_temp[book_temp['Book-Title']==pt.index[i[0]]]['Book-Author'].values))
        lists.extend(list(book_temp[book_temp['Book-Title']==pt.index[i[0]]]['Image-URL-M'].values))
        #lists.append(book[book['Book-Title']==pt.index[i[0]]]['avg_ratings'].values)
        recommend.append(lists)
    return recommend
```

```python
recommend(1984)
```
